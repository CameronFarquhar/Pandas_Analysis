```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

df = pd.DataFrame()
```

## Player Count

* Display the total number of players


```python
#create a list describing the total amount of players and find the length of that list
unique_players = purchase_data["SN"].unique()
total_num = [{"Total Players":len(unique_players)}]
total_num_df = pd.DataFrame(total_num)
total_num_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# determine how many times each item was purchased
item_count = purchase_data["Item ID"].value_counts()
 
item_count_len = len(item_count)

# find average price, number of purchases, and total amount of revenue
av_price = purchase_data["Price"].mean()

num_purch = len(purchase_data["Item ID"])

total_rev = sum(purchase_data["Price"])

# Put it all into a dictionary
analysis_dict = [{"Number of Unique Items": item_count_len,
                 "Average Price": av_price,
                 "Number of Purchases": num_purch,
                 "Total Revenue":total_rev}]

# create a data frame
analysis_df = pd.DataFrame(analysis_dict)

# add formating
analysis_df["Average Price"] = analysis_df["Average Price"].map('${:,.2f}'.format)

analysis_df["Total Revenue"] = analysis_df["Total Revenue"].map('${:,.2f}'.format)

analysis_df
```

## Purchasing Analysis (Gender)

```python
# filter the raw data by gender
group_gender = purchase_data.groupby(["Gender"])

#find the total amount of users and how many of which identify as Male, Female, or Other
name_unique = group_gender["SN"].unique()

female = len(name_unique["Female"])

male = len(name_unique["Male"])

other = len(name_unique["Other / Non-Disclosed"])
#find the total in order to determine the average
total_gender = (female + male + other)

gender_demo_dict = {
    "Total Count": [male, female, other],
    "Percentage of Players": [(male/total_gender*100), (female/total_gender*100), (other/total_gender*100)]
}

gender_demo_df = pd.DataFrame(gender_demo_dict, index=['Male','Female','Other / Non-Disclosed'])  

gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].map('{:,.2f}%'.format)

gender_demo_df
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#determine the total amount of each gender's purchase
gender_count = purchase_data["Gender"].value_counts()

#determine the average purchase price per gender
gend_av_purchase = group_gender["Price"].mean()

#determine total purchases per gender
gend_sum_purchase = group_gender["Price"].sum()

gender_purch_dict = {
    "Purchase Count": [gender_count["Female"], gender_count["Male"], gender_count["Other / Non-Disclosed"]],
    "Average Purchase Price": [gend_av_purchase["Female"], gend_av_purchase["Male"], gend_av_purchase["Other / Non-Disclosed"]],
    "Total Purchase Value": [gend_sum_purchase["Female"], gend_sum_purchase["Male"], gend_sum_purchase["Other / Non-Disclosed"]],
    "Avg Total Purchase per Person": [(gend_sum_purchase["Female"]/female), (gend_sum_purchase["Male"]/male), (gend_sum_purchase["Other / Non-Disclosed"]/other)]
}

gender_analysis_df = pd.DataFrame(gender_purch_dict, index=['Female','Male','Other / Non-Disclosed']) 

gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].map('${:,.2f}'.format)

gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map('${:,.2f}'.format)

gender_analysis_df["Avg Total Purchase per Person"] = gender_analysis_df["Avg Total Purchase per Person"].map('${:,.2f}'.format)

gender_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#bins and labels
bins = [0,9,14,19,24,29,34,39,1000]

labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

purchase_data["Total Count"] = pd.cut(purchase_data["Age"], bins=bins, labels=labels)

#get the total number of individual users
sn_age = len(purchase_data["SN"].unique())

#GroupBy bins
age_group = purchase_data.groupby(["Total Count"])

#create a list of names in each age group to determine the amount of individuals in the age group
age_group_count = age_group["SN"].unique()

age_group_dict = {
    "Total Count":[len(age_group_count["<10"]), len(age_group_count['10-14']), len(age_group_count['15-19']), len(age_group_count['20-24']), len(age_group_count['25-29']), len(age_group_count['30-34']), len(age_group_count['35-39']), len(age_group_count['40+'])],
    "Percentage of Players": [(len(age_group_count["<10"])/sn_age*100), (len(age_group_count['10-14'])/sn_age*100), (len(age_group_count['15-19'])/sn_age*100), len(age_group_count['20-24'])/sn_age*100, (len(age_group_count['25-29'])/sn_age*100), (len(age_group_count['30-34'])/sn_age*100), (len(age_group_count['35-39'])/sn_age*100), (len(age_group_count['40+'])/sn_age*100)]
}
age_group_df = pd.DataFrame(age_group_dict, index=labels)

age_group_df["Percentage of Players"] = age_group_df["Percentage of Players"].map('{:,.2f}%'.format)

age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins_group = purchase_data.groupby(["Total Count"])

#this shows the total number of puchases in each age group
bins_group_count = bins_group["Price"].count()

#average purchase price
bins_group_average = bins_group["Price"].mean()

#total purchase value
bins_group_sum = bins_group["Price"].sum()

#create a dictionary

bins_group_dict = {
    "Purchase Count":bins_group_count,
    "Average Purchase Price":bins_group_average,
    "Total Purchase Value":bins_group_sum,
     "Avg Total Purchase per Person": [bins_group_sum["<10"]/len(age_group_count["<10"]), bins_group_sum['10-14']/len(age_group_count['10-14']), bins_group_sum['15-19']/len(age_group_count['15-19']), bins_group_sum['20-24']/len(age_group_count['20-24']), bins_group_sum['25-29']/len(age_group_count['25-29']), bins_group_sum['30-34']/len(age_group_count['30-34']), bins_group_sum['35-39']/len(age_group_count['35-39']), bins_group_sum['40+']/len(age_group_count['40+'])]
}
bins_group_df = pd.DataFrame(bins_group_dict)

bins_group_df["Average Purchase Price"] = bins_group_df["Average Purchase Price"].map('${:,.2f}'.format)

bins_group_df["Total Purchase Value"] = bins_group_df["Total Purchase Value"].map('${:,.2f}'.format)

bins_group_df["Avg Total Purchase per Person"] = bins_group_df["Avg Total Purchase per Person"].map('${:,.2f}'.format)

bins_group_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
    ##group by Screen Name
sn_group = purchase_data.groupby(["SN"])

    ##find the average number of purchases from each user
sn_mean = sn_group["Price"].mean()

    ##find the total number of purchases from each user
sn_count = sn_group["Price"].count()

    ##Create a variable to multiply purchase count by average to get total
sn_total = sn_mean*sn_count

sn_group_dict = {
    "Purchase Count":sn_count,
    "Average Purchase Price":sn_mean,
    "Total Purchase Value":sn_total,
}
sn_group_df = pd.DataFrame(sn_group_dict)


    ##Now sort the data from largest to smallest total purchase value

sn_sort = sn_group_df.sort_values("Total Purchase Value", ascending=False).head(5)

sn_sort["Average Purchase Price"] = sn_sort["Average Purchase Price"].map('${:,.2f}'.format)

sn_sort["Total Purchase Value"] = sn_sort["Total Purchase Value"].map('${:,.2f}'.format)

sn_sort
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
id_name_group = purchase_data.groupby(["Item ID","Item Name"])

#doing this one without variables because it is fewer lines of code
id_name_dict = {
    "Purchase Count":id_name_group["Item ID"].count(),
    "Item Price":id_name_group['Price'].sum()/id_name_group["Item ID"].count(),
    "Total Purchase Value":id_name_group['Price'].sum()
}

id_name_df = pd.DataFrame(id_name_dict)

id_name_sort = id_name_df.sort_values("Purchase Count", ascending=False).head(5)

id_name_sort["Item Price"] = id_name_sort["Item Price"].map('${:,.2f}'.format)

id_name_sort["Total Purchase Value"] = id_name_sort["Total Purchase Value"].map('${:,.2f}'.format)

id_name_sort
```
